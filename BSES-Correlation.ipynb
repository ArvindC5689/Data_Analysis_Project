# Correlation Coefficients for BSES

```python
import numpy as np 
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
```

### Formula for BSES:

#### BSES = Situation Factor * (RBI Efficiency + Event Score)

```python
local_path = 'BSES.csv'


file_path = local_path 


dfTW= pd.read_csv(file_path)


dfTW.head()
```

```python
dfTW2 = dfTW.loc[dfTW['PA']>500]
dfTW2.head()
```

## BSES vs. PA

```python
print("Correlation =", dfTW2['BSES'].corr(dfTW2['PA'], method='pearson'))
fig, axes = plt.subplots(figsize=(7,7));
axes.scatter(dfTW2['BSES'], dfTW2['PA'], color = "Blue");
```

## BSES vs. RBIs

```python
print("Correlation =", dfTW2['BSES'].corr(dfTW2['RBI'], method='pearson'))
fig, axes = plt.subplots(figsize=(7,7));
axes.scatter(dfTW2['BSES'], dfTW2['RBI'], color = "Purple");
```
